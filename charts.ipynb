```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
gr = '#333333'
data = np.append(np.arange(230,234),np.arange(241,247))
data = np.append(data,np.arange(253,259))
bdf = pd.read_csv(r'D:\Mestrado\Imagens\Antigas\Comparison\csv\{}.csv'.format(229))
bdf.drop('Unnamed: 0',axis=1,inplace=True)
for i in data:
    rdf = pd.read_csv(r'D:\Mestrado\Imagens\Antigas\Comparison\csv\{}.csv'.format(i))
    rdf.drop('Unnamed: 0',axis=1,inplace=True)
    bdf = pd.concat([bdf,rdf])

bdf.set_index('num',inplace=True)
bdf.index.name = None

jg_30_alta = pd.read_csv(r'D:\Mestrado\Old\mean.csv').loc[29:,'JG_out_adm']
jg_60_alta = pd.read_csv(r'D:\Mestrado\Old\mean.csv').loc[23:28,'JG_out_adm']
jg_90_alta = pd.read_csv(r'D:\Mestrado\Old\mean.csv').loc[18:22,'JG_out_adm']
```

```python
fig = plt.figure(figsize=(14,6/np.sqrt(2)))
plt.suptitle('30Â°')
ax1 = fig.add_subplot(121)
ax1.plot(jg_30_alta,bdf.iloc[11:,0].as_matrix(),
         'o-',label='Interface detection',color=gr)
ax1.plot(jg_30_alta,bdf.iloc[11:,1].as_matrix(),
         'o-',label='Image superposition',color=gr,mec=gr,mfc='w')
#plt.plot(bdf.iloc[:,0],bdf.iloc[:,2] - bdf.iloc[:,1],color=gr)
ax1.set_ylim([0,50])
ax1.set_ylabel('f (Hz)')
ax1.set_xlabel(r'$j_{g,o}^*$ (-)')
ax1.legend()
plt.grid('on');
ax2 = fig.add_subplot(122)
ax2.plot(jg_30_alta,bdf.iloc[11:,2].as_matrix(),
         'o-',label='Interface detection',color=gr)
ax2.plot(jg_30_alta,bdf.iloc[11:,3].as_matrix(),
         'o-',label='Image superposition',color=gr,mec=gr,mfc='w')
#plt.plot(bdf.iloc[:,0],bdf.iloc[:,2] - bdf.iloc[:,1],color=gr)
ax2.set_ylim([300,600])
ax2.set_ylabel('v (mm/s)')
ax2.set_xlabel(r'$j_{g,o}^*$ (-)')
ax2.legend()
plt.grid('on');
```
