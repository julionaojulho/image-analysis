```python
def hide_code_in_slideshow():
    import os
    uid = os.urandom(8).hex()
    html = """<div id="%s"></div>
    <script type="text/javascript">
        $(function(){
            var p = $("#%s");
            if (p.length==0) return;

            while (!p.hasClass("cell")) {
                p=p.parent();

                if (p.prop("tagName") =="body") return;
            }
            var cell = p;
            cell.find(".input").addClass("hide-in-slideshow")
        });
    </script>""" % (uid, uid)
    display_html(html, raw=True)
```


# Interface detection vs Image superposition

### importing modules

- numpy for numerical programming
- correlate function to perform the cross correlation of signals
- matplotlib for charts
- pandas for exporting data as csv
- homemade functions to load and format the images
- IPython methods to display content

```python
import numpy as np
from scipy.signal import correlate
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
from my_modules.inthei import *
from IPython.display import Image, SVG, display, HTML, IFrame, display_html
from timeit import default_timer as timer
```

## Interface detection

```python
num = 233
gr = '#333333'
pics = np.linspace(1,9e3,9e3)
location = r'\Antigas\Alta\90\{}\Imagem '.format(num)
ROI = (0,792,1,110)
start = timer()
iarray = im_start(pics,path=location,box=ROI)
dt = timer() - start
print(dt)
```

```python
start = timer()
iarray_b = bg_removal(iarray)
dt = timer() - start
print(dt)
```

```python
start = timer()
iarray_p = np.array([im_proc(i) for i in iarray_b])
scale = ROI[3]-ROI[2]
dt = timer() - start
print(dt)
```

```python
start = timer()
h1 = np.zeros((iarray_p.shape[0],iarray_p.shape[1]))
for j,ii in enumerate(iarray_p):
    iarray_D, iarray_h = interface_height(ii,scale)
    h1[j] = iarray_h
dt = timer() - start
print(dt)
```

###  Time and space signals

```python
hide_code_in_slideshow()

fig = plt.figure(figsize=(6,4/np.sqrt(2)))
ax1 = fig.add_subplot(211)
ax1.plot(scale-h1[:,100],color=gr)
ax1.set_title('Time')
ax1.set_xlim([0,9e3])
ax1.set_ylim([0,1.2*max(scale-h1[:,100])])
ax2 = fig.add_subplot(212)
ax2.plot(scale-h1[100,:],color=gr)
ax2.set_title('Space')
ax2.set_xlim([0,792])
ax2.set_ylim([0,1.2*max(scale-h1[1,:])])
plt.tight_layout()
fig.savefig(r'D:\Mestrado\Imagens\Antigas\Comparison\{} - sig_t1.png'.format(num))
```

### Fourier analysis - time signal and its spectrum

```python
hide_code_in_slideshow()

sig = scale-h1[:,100]
fs = 3000
t = np.linspace(0,1000/fs,len(sig))
Y = np.fft.fft(sig - np.mean(sig),norm='ortho')
Y = 2 * Y / len(Y)
Y[0] = Y[0] / 2
f = np.fft.fftfreq(len(Y),d=1/fs)

fig = plt.figure(figsize=(6,4/np.sqrt(2)))
ax1 = fig.add_subplot(211)
ax1.plot(t,sig,color=gr)
ax1.set_xlim([0,0.33])
ax1.set_ylim([0,1.2*max(scale-h1[:,100])])
ax2 = fig.add_subplot(212)
ax2.plot(f[0:int(len(Y)/2)],np.abs(Y[0:int(len(Y)/2)])**2,'o-',
        color=gr,label='df = {:.2f} Hz'.format(f[1]-f[0]))
ax2.set_xlim([0, 200])
ax2.set_ylim([0, 1.2*np.max(np.abs(Y[0:int(len(Y)/2)])**2)])
plt.ticklabel_format(style='sci', axis='y', scilimits=(0,4))
ax2.yaxis.major.formatter._useMathText = True
ax2.legend(loc='best',numpoints=1,fontsize=10)
plt.tight_layout()
fig.savefig(r'D:\Mestrado\Imagens\Antigas\Comparison\{} - four_t1.png'.format(num))
```

```python
fs = 3000
freq = []
for i,ts in enumerate(h1.T):
    ts_flip = np.mean(ts) - ts
    Y = np.fft.fft(ts_flip,norm='ortho')
    Y = 2 * Y / len(Y)
    Y[0] = Y[0] / 2
    f = np.fft.fftfreq(len(Y),d=1/fs)
    idx = np.argmax(np.abs(Y[0:int(len(Y)/2)])**2)
    freq.append(f[idx])
print('Frequncia mdia: {:.2f} Hz'.format(np.mean(freq)))
```

## Histograms

### Frequency and velocity

```python
b = []
for i in range(np.shape(h1.T)[0]-50):
    b.append(np.argmax(correlate(5*(scale-h1.T[i,:9000])/scale,
                                 5*(scale-h1.T[i+50,:9000])/scale,mode='same')))
b = [i for i in b if i != 4500]
v = 50 / (np.array(b) - 4500) * (5 / scale) * fs
print(np.mean(v))
```

```python
hide_code_in_slideshow()

fig = plt.figure(figsize=(12,4))
ax1 = fig.add_subplot(121)
ax1.hist(freq,bins=10,orientation='horizontal',
         histtype='bar',color=gr,normed=False,
        label='Mean frequency: {:.2f} Hz'.format(np.mean(freq)))
ax1.set_ylabel('Frequency (Hz)')
ax1.set_xlabel('Occurrences (-)')
plt.legend();

ax2 = fig.add_subplot(122)
ax2.hist(v,bins=10,
        orientation='horizontal',color=gr,
        label='Mean velocity: {:.2f} mm/s'.format(np.mean(v)))
ax2.set_ylabel('Velocity (mm/s)')
ax2.set_xlabel('Occurrences (-)')
plt.legend();
fig.savefig(r'D:\Mestrado\Imagens\Antigas\Comparison\{} - hist_t1.png'.format(num));

t1_f,t1_v = np.mean(freq),np.mean(v)
```

##  Image superposition

Adjustments are required for "ROI2" - film thickness increases, therefore the interface is displaced

```python
ROI2 = (0,792,85,86)
imgs = np.zeros((1,792),dtype="int32")
for i in np.linspace(1,9000,9000):
    a = imread_x(i,path=location,box=ROI2).T
    imgs = np.concatenate((imgs,a))
imgs = imgs[1:,:]
```

###  Time and space signals

```python
hide_code_in_slideshow()
fig = plt.figure(figsize=(6,4/np.sqrt(2)))
ax1 = fig.add_subplot(211)
ax1.plot(imgs.T[100],color=gr)
ax1.set_title('Time')
ax1.set_xlim([0,9e3])
ax1.set_ylim([0,1.2])
ax2 = fig.add_subplot(212)
ax2.plot(imgs.T[:,100],color=gr)
ax2.set_title('Space')
ax2.set_xlim([0,792])
ax2.set_ylim([0,1.2])
plt.tight_layout()
fig.savefig(r'D:\Mestrado\Imagens\Antigas\Comparison\{} - sig_t2.png'.format(num))
```

### Fourier analysis - time signal and its spectrum

```python
hide_code_in_slideshow()
sig = imgs.T[100] - np.mean(imgs.T[100])
fs = 3000
t = np.linspace(0,1000/fs,len(sig))
Y = np.fft.fft(sig,norm='ortho')
Y = 2 * Y / len(Y)
Y[0] = Y[0] / 2
f = np.fft.fftfreq(len(Y),d=1/fs)
fig = plt.figure(figsize=(6,4/np.sqrt(2)))
ax1 = fig.add_subplot(211)
ax1.plot(t,sig,color=gr)
ax1.set_xlim([0,0.33])
ax1.set_ylim([0,1.2])
ax2 = fig.add_subplot(212)
ax2.plot(f[0:int(len(Y)/2)],np.abs(Y[0:int(len(Y)/2)])**2,'o-',
        color=gr,label='df = {:.2f} Hz'.format(f[1]-f[0]))
ax2.set_xlim([0, 200])
ax2.set_ylim([0, 1.2*np.max(np.abs(Y[0:int(len(Y)/2)])**2)])
plt.ticklabel_format(style='sci', axis='y', scilimits=(0,4))
ax2.yaxis.major.formatter._useMathText = True
ax2.legend(loc='best',numpoints=1,fontsize=10)
plt.tight_layout()
fig.savefig(r'D:\Mestrado\Imagens\Antigas\Comparison\{} - four_t2.png'.format(num))
```

```python
freq = []
for i,ts in enumerate(imgs.T):
    ts_flip = ts - np.mean(ts)
    Y = np.fft.fft(ts_flip,norm='ortho')
    Y = 2 * Y / len(Y)
    Y[0] = Y[0] / 2
    f = np.fft.fftfreq(len(Y),d=1/fs)
    idx = np.argmax(np.abs(Y[0:int(len(Y)/2)])**2)
    freq.append(f[idx])
print('Frequncia mdia: {:.2f} Hz'.format(np.mean(freq)))
```

### Histogram
